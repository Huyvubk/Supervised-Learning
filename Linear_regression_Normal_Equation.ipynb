```python
# Normal Equation
```

```python
import numpy as np
import tensorflow as tf

### Data preparation
data = np.genfromtxt('/content/drive/MyDrive/Data/advertising.csv', delimiter=',',skip_header= True)

x = data[:,:-1]
y = data[:,-1]

```

```python
x.shape
```

```python
X = np.concatenate((x, np.ones((len(x),1))), axis = 1)
```

```python
X.shape,y.shape
```

```python
def find_theta(X,y):
    return np.dot(np.dot(np.linalg.inv(np.dot(X.T,X)), X.T), y)
```

```python
w = find_theta(X,y)
```

```python
y_predict = X.dot(w)
```

```python
np.abs(y_predict - y).mean()
```

```python
for i,j in zip(y_predict , y):
    print( f"{i} :  {j}")
```
